```python
fastfood = pd.read_csv("fastfoodalleghenycountyupdatexy2.csv")
fastfood.head(10)
```

```python
zipcode = fastfood.groupby('ZIP_Code').count()
zipcode_sorted = zipcode.sort_values(by=['Name'], ascending = False)
zipcode_sorted.head(20)
```

```python
zipcode = fastfood.groupby('ZIP_Code').count()
zipcode.sort_values(by=['Name'], ascending = False)
```

```python
from shapely.geometry import Point

# combine lat and lon column to a shapely Point() object
zipcode['geometry'] = zipcode.apply(lambda x: Point((float(x.Lon), float(x.Lat))), axis=1)
```

```python
import geopandas
zipcode = geopandas.GeoDataFrame(zipcode, geometry='geometry')
```

```python
zipcode.to_file('MyGeometries.shp', driver='ESRI Shapefile')
```
